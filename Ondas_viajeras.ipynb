#Gráfique la función

$$f(x') = e^{-\frac{(x'-0.5)^2}{0.01}}$$

#y suponga que esta representa una perturbación en una cuerda

Importamos las librerias

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definimos la función

```python
# Definir la función original
def f(x_prime):
    return np.exp(-(x_prime - 0.5)**2 / 0.01)
```

Mejoramos el panel de la gráfica

```python
x = np.linspace(0, 1, 500)
c = 0.1
t_values = [0, 1, 2, 3]
plt.figure(figsize=(15, 10))
```

Graficamos la función

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x_prime):
  return np.exp(-(x_prime - 0.5)**2 / 0.01)

x_prime = np.linspace(0, 1, 100)
y = f(x_prime)

plt.plot(x_prime, y)
plt.title("Turbación en una cuerda")
plt.xlabel("x'")
plt.ylabel("f(x')")
plt.grid(True)
plt.show()
```

#a) Mostrar que al hacer el cambio de variable $x'=x-ct$ la perturbación se desplaza a la derecha

Definimos los distintos valores para $x'$

```python
plt.subplot(3, 1, 1)
```

Hacemos el cambio de variable, pra cada tiempo de $t$. Graficamos la función

```python
for t in t_values:
    x_prime = x - c*t
    plt.plot(x, f(x_prime), label=f't = {t}')
plt.title('(a) Perturbación desplazándose a la derecha (x\' = x - ct)')
plt.xlabel('x')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
```

#b) Mostrar que al hacer el cambio de variable $x'=x+ct$, la perturbación se desplaza hacia izquierda

Para los distintos valores de $x'$

```python
plt.subplot(3, 1, 1)
```

Hacemos el cambio de variable y graficamos

```python
for t in t_values:
    x_prime = x + c*t
    plt.plot(x, f(x_prime), label=f't = {t}')
plt.title('(b) Perturbación desplazándose a la izquierda (x\' = x + ct)')
plt.xlabel('x')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
```

#Graficar la superposión de las ondas viajeras de los incisos a) y b)

Graficamos el panel

```python
plt.subplot(3, 1, 3)
```

Para cada valor de $t$ definimos las funciones y las graficamos

```python
for t in t_values:
    onda_derecha = f(x - c*t)
    onda_izquierda = f(x + c*t)
    superposicion = onda_derecha + onda_izquierda
    plt.plot(x, superposicion, label=f't = {t}')
    plt.title('(c) Superposición de ambas perturbaciones viajeras')
plt.xlabel('x')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
