Grafica de una parabola cuya ecuacion es $f(x)=x^2$



Importar las librerias que se necesitan para graficar


```python
# Codigo: graficar y=x^2

#Importar librerias
import matplotlib.pyplot as plt
import numpy as np

# Generar x valores
x = np.linspace(-5, 5, 400)

# Calcular y valores
y = x**2

# Crear plano
plt.plot(x, y, color='blue')

# Etiquetas y titulo
plt.xlabel("x", color='red')
plt.ylabel("y", color='red')
plt.title("Grafica de y = xÂ²", color='red')

# Mostrar la grafica
plt.grid(True)
plt.show()
```
