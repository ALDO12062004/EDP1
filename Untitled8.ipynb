Grafica de una parabola cuya ecuacion es $f(x)=x^2$

Importar las librerias que se necesitan para graficar

```python
# Codigo: graficar y=x^2

import matplotlib.pyplot as plt
import numpy as np

# Generar x valores
x = np.linspace(-5, 5, 400)

# Calcular y valores
y = x**2

# Crear la grafica
plt.plot(x, y)

# Etiquetas y titulo
plt.xlabel("x")
plt.ylabel("y ")
plt.title("Grafica de y = x²")

# Mostrar la grafica
plt.grid(True)
plt.show()
```
