```python
# Codigo: graficar y=x^2

import matplotlib.pyplot as plt
import numpy as np

# Generar x valores
x = np.linspace(-5, 5, 400)

# Calculate y values
y = x**2

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("x")
plt.ylabel("y = x²")
plt.title("Graph of y = x²")

# Display the plot
plt.grid(True)
plt.show()
```
